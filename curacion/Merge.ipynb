### Una alternativa para unir las bases de datos clínicos usando DataFrame.merge()

### Importamos paquetes

```python
import os
import pandas as pd
```

### Cargamos los datasets clínicos

```python
data_dir = os.path.join('..', 'data')

dataset_con_paciente = os.path.join(data_dir, 'turnos_con_paciente_2018-2019_despersonalizado.csv')
dataset_sin_paciente = os.path.join(data_dir, 'turnos_sin_paciente_2018-2019_despersonalizado.csv')

dfp = pd.read_csv(dataset_con_paciente, sep=';')
dfsp = pd.read_csv(dataset_sin_paciente, sep=';')
```

```python
## Renombramos columnas
to_rename = {
    'Centro de atención': 'CentroDeAtencion',
    'Prestación asignada': 'Prestacion',
    'Estado del turno': 'EstadoDelTurno',
    'Nombre del canal': 'Canal',
    'Fecha del turno': 'FechaDelTurno',
    'Hora del turno': 'HoraDelTurno',
    'Es sobre turno': 'EsSobreTurno',
    'Tipo de turno asignado': 'TipoDeTurno'
}
dfsp2 = dfsp.rename(columns=to_rename)

## Parseamos fechas
dfsp2.FechaDelTurno = dfsp2.FechaDelTurno + ' ' + dfsp2.HoraDelTurno
dfsp2.FechaDelTurno = pd.to_datetime(dfsp2.FechaDelTurno)
dfsp2 = dfsp2.drop(columns=['HoraDelTurno'])

## Renombramos columnas
to_rename = {
    'CentroAtencion_Nombre': 'CentroDeAtencion',
    'Servicio_Nombre': 'Servicio',
    'TurnoEstado_Nombre': 'EstadoDelTurno',
    'FechaTurno': 'FechaDelTurno',
    'TipoTurno': 'TipoDeTurno'
}
dfp2 = dfp.rename(columns=to_rename)

## Parseamos fechas
dfp2.FechaDelTurno = pd.to_datetime(dfp2.FechaDelTurno)
```

```python
cols = [
    'CentroDeAtencion', 'Servicio', 'Recurso', 'FechaDelTurno', 
    'EstadoDelTurno', 'Prestacion', 'TipoDeTurno'
]
df = dfp2.merge(dfsp2, on=cols)
df.head()
```
